## View Data

The `data` folder contains the sample training data for this workshop. It is a series of simple cartoons of a young boy Malcom and his cute dog Ben.

You will use this data to train a fine-tuned LoRA adapter for Stable Diffusion 3.5 Medium, which will return images of the same character and art style. While the dataset only has 11 low quality samples, it highlights what is possible with better/more data and longer training cycles.

```python
from datasets import load_from_disk
dataset = load_from_disk("data")

print(f"items in training dataset: {len(dataset["train"])}")
```

Each sample consists of a `caption` used to drive image generation and an `image` that is used to determine training loss.

You can increment the `index` value to see other entries. 

```python
from PIL import Image

index = 1

print(dataset['train'][index]['caption'])
dataset['train'][index]['image'].show()
```
